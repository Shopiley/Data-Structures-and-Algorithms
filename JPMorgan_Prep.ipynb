### Q.1 Delete Anagrams

An anagram of a word, is another word that is of the same length as the original word but with a change in the order the characters.
You are given an array of Strings. You read the array from left to right, if you read a word that is an anagram of any previous word, delete the current word. Return the array with all anagrams deleted by following the above mentioned method. Return the array in ascending alphabetical order.

eg. arr = ["code" , "dcoe" , "timer", "timerr", "remit"]
answer = ["code", "timer", "timerr"]

Declare a set(keep it unordered for better time complexity).

Start traversing from left of the array.
For each string in the array
2.1 sort the string.
2.2 if the string is not present in the set, put it in answer array, else continue.
Put the current sorted string in the set.
At last, return the answer array.

```python
def deleteAnagrams(arr):
    a = list(arr[0])
    a.sort()
    a = ''.join(a)
    hashSet = set()
    hashSet.add(a)
    result = [arr[0]]

    for i in range(1, len(arr)):
        sorted = list(arr[i])
        sorted.sort()   #n^2logn 
        sorted = ''.join(sorted)
        if sorted not in hashSet:
            hashSet.add(sorted)
            result.append(arr[i])
        
    result.sort()
    return result

arr = [ "timer", "timerr", "remit", "code" , "dcoe"]
print(deleteAnagrams(arr))

```

```python
def deleteAnagrams(arr):
    hashSet = set()
    hashSet.add(''.join(sorted(arr[0])))
    result = [arr[0]]

    for i in range(1, len(arr)):
        sorted_word = ''.join(sorted(arr[i]))

        if sorted_word not in hashSet:
            hashSet.add(sorted_word)
            result.append(arr[i])
        
    result.sort()
    return result

arr = [ "timer", "timerr", "remit", "code" , "dcoe"]
print(deleteAnagrams(arr))
```

### Q.2 Match Expression with minimum number of operations

An expression is said to be matched if for every '>' there is a corresponding '<' to the left it. Example of a matched expression is <<><>>.
Say you have an unmatched expression eg <>>, then can you convert it to a matched expression if the only operation you can do is replace a '>' with a '<>'.
Two arrays were given: expressionsArray, maxReplacements both were of the same length.
We had to iterate through the expressionsArray and check if we can return a balanced expression (say for expressionsArray[i] if we are allowed to execute the above mentioned operation only maxReplacements[i] number of times.
Return a boolean array indicating if it is possible.

eg)
expressionsArr = ["<<>>", "<>", "<><>", ">>", "<<>" , "><><"]
maxReplacements = [0, 1, 2, 2, 2, 2]
Answer = [1, 1, 1, 1, 0, 0]

```python
def matchExpression(expr, maxRep):

    stack = [expr[0]]
    replacements, remainder = 0, 0

    for i in range(1, len(expr)):
        if stack and stack[-1] == "<" and expr[i] == ">":
            stack.pop()
        else:
            stack.append(expr[i])
    
    for brac in stack:
        if brac == ">":
            replacements += 1
        else:
            remainder += 1


    if replacements > maxRep or remainder:
        return 0
    return 1

"""
time: O(n)
space: O(n)
"""
# print(matchExpression("<<>", 1))
# print(matchExpression("<<><><<>>", 0))
print(matchExpression("<>>>", 2))
print(matchExpression("<>>>>", 2))
# print(matchExpression("<>><", 1))

expressionsArr = ["<<>>", "<>", "<><>", ">>", "<<>" , "><><"]
maxReplacements = [0, 1, 2, 2, 2, 2]

print([matchExpression(expressionsArr[i], maxReplacements[i]) for i in range(len(expressionsArr))])
```

### DNA String



```python
def DNAString(s: str):
    stack = list(s)
    reversed = []

    while stack:
        reversed.append(stack.pop())
    
    for i in range(len(reversed)):
        if reversed[i] == "A": reversed[i] = "T" 
        elif reversed[i] == "T": reversed[i] = "A"
        elif reversed[i] == "C": reversed[i] = "G"
        elif reversed[i] == "G": reversed[i] = "C"

    return ''.join(reversed)

print(DNAString("GTCAG"))

"""
time complexity: O(n)
space complexity: O(n)
"""
```

### Missing words


```python
def missingWords(s, t):
    s = s.split()
    t = t.split()
    ans = []

    for word in s:
        if word not in t:
            ans.append(word)

    return ans

s = "I like cheese"
t = "like"

print(missingWords(s, t))
```

```python
def missingWords(s, t):
    s = set(s.split())
    t = set(t.split())

    ans = []

    for word in s:
        if word not in t: # TC: O(1) because it's a set
            ans.append(word)

    return ans

s = "I like cheese"
t = "like"

print(missingWords(s, t))
```

```python


def minimalOperations(word):
    count = 0
    i = 1
    while i < len(word):
        if word[i-1] == word[i]:
            i += 1
            count += 1
        i += 1
    return count
    

words = ["add", "booook", "break", "aduunuu"]
for word in words:
    print(minimalOperations(word))
```

### AdRotation

```python
def changeAds(base10):
    binary = []
    while base10:
        temp = base10//2
        rem = base10 % 2
        binary.append(rem)
        base10 = temp
    binary.reverse()
    binaryComp = toComplement(binary)
    binaryComp = int(''.join(map(str, binaryComp)))
    return toBase10(str(binaryComp))
        

def toComplement(binary):
    for i in range(len(binary)):
        if binary[i] == 1:
            binary[i] = 0
        elif binary[i] == 0:
            binary[i] = 1
    return binary


def toBase10(num):
    num = list(num)
    idx = 0
    exp = len(num) - 1
    ans = 0

    while idx < len(num) and exp >= 0:
        ans += int(num[idx])*(2**exp)
        idx += 1
        exp -= 1
    return ans


print(changeAds(100))
```
